```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
# AI algorithms can not handle categorical data, since the data set contains categorical values in fewer columns like sg,rbc, pc, pcc, ba, htn, etc.,  column contains 
# we need to convert the categorical value into numerical values
# using below method.
dataset=pd.get_dummies(dataset,drop_first=True) # Removing the repeated value column
```

```python
dataset # Once the categorical columns converted into numerical column values
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']] # separating the input values
dep=dataset['classification_yes'] # separating the output values
```

```python
indep
```

```python
dep
```

```python
dep=dataset[['classification_yes']]
```

```python
dep
```

```python
#print("Multi line comment working...")



#########################   SVC GRID SEARCH NO test and training data split ######################################

# Consolidating the playable parameters probability and also considering the cross validation
# Considering different portions of a data set as a test and traint set
from sklearn.model_selection import GridSearchCV 
from sklearn.svm import SVC
param_grid = {'C': [0.1, 1, 10],  # Regularization parameter
    'kernel': ['linear', 'rbf'],  # Kernel type
    'gamma': [0.1, 1, 10] } # Kernel coefficient for 'rbf'} # Giving all possible HTPs
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
# refit = True means consider the best result model as a result model
# refit = False means consider the last combination model as a result model




'''
#########################   Decision Tree GRID SEARCH NO test and training data split ######################################

# Consolidating the playable parameters probability and also considering the cross validation
# Considering different portions of a data set as a test and traint set
from sklearn.model_selection import GridSearchCV 
from sklearn.tree import DecisionTreeClassifier

#Giving all possible HTPs
param_grid = {'criterion':['gini', 'entropy'], 'splitter':['best', 'random']}


grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
# refit = True means consider the best result model as a result model
# refit = False means consider the last combination model as a result model
'''


'''
#########################   Random Forest GRID SEARCH NO test and training data split ######################################

# Consolidating the playable parameters probability and also considering the cross validation
# Considering different portions of a data set as a test and traint set
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestClassifier

#Giving all possible HTPs
param_grid = {'criterion':['gini','entropy'],
'max_features': ['sqrt','log2'],
'n_estimators':[10,100]} # Giving all possible HTPs
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# refit = True means consider the best result model as a result model
# refit = False means consider the last combination model as a result model
'''
```

```python
# fitting the model for grid search
grid.fit(indep, dep) # Passing whole dataset input and output variable
```

```python
# print best parameter after tuning
print(grid.best_params_)
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```
